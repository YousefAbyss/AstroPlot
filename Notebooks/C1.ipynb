```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science'])
```

```python
mpl.rcParams['text.latex.preamble'] = r'\boldmath'

mpl.rcParams['xtick.major.size'] = 6
mpl.rcParams['xtick.major.width'] = 2
mpl.rcParams['xtick.minor.size'] = 3
mpl.rcParams['xtick.minor.width'] = 1
mpl.rcParams['xtick.labelsize'] = 13
mpl.rcParams['xtick.direction'] = 'inout'

mpl.rcParams['ytick.major.size'] = 6
mpl.rcParams['ytick.major.width'] = 2
mpl.rcParams['ytick.minor.size'] = 3
mpl.rcParams['ytick.minor.width'] = 1
mpl.rcParams['ytick.labelsize'] = 13
mpl.rcParams['ytick.direction'] = 'inout'

mpl.rcParams['axes.linewidth'] = 1.2
mpl.rcParams['axes.labelsize'] = 15
mpl.rcParams['axes.labelweight'] = 'bold'

mpl.rcParams['figure.titlesize'] = 14
```

```python
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(6, 5), sharex=True, sharey=True)

x = np.linspace(0, 10, 500)

y = np.linspace(0, 10, 500)

z = np.zeros([500, 500])

for i in range(500):
    for j in range(500):
        z[i, j] = x[i] + np.sin(y[j])

ax[0, 0].pcolormesh(x, y, z, cmap='viridis')

for i in range(500):
    for j in range(500):
        z[i, j] = np.cos(x[i]) + y[j]

ax[1, 0].pcolormesh(x, y, z, cmap='viridis')

for i in range(500):
    for j in range(500):
        z[i, j] = np.sqrt(x[i] + y[j] ** 2)

ax[0, 1].pcolormesh(x, y, z, cmap='viridis')

for i in range(500):
    for j in range(500):
        z[i, j] = 9 * np.sin(x[i] + y[j])

pcm1 = ax[1, 1].pcolormesh(x, y, z, cmap='viridis', vmin=0, vmax=10)

cbar = fig.colorbar(pcm1, ax=ax, orientation='vertical', anchor=(1.7, 1))
cbar.set_label(r'$z$', labelpad=0)

ax[0, 0].set_ylabel(r'$y$')
ax[1, 0].set_ylabel(r'$y$')
ax[1, 0].set_xlabel(r'$x$')
ax[1, 1].set_xlabel(r'$x$')

ax[0, 0].set_xticks([2 + i * 2 for i in range(5)])
ax[0, 0].set_yticks([2 + i * 2 for i in range(5)])

plt.subplots_adjust(wspace=0, hspace=0)
# plt.savefig('C1.png', dpi=150)

plt.show()
```
